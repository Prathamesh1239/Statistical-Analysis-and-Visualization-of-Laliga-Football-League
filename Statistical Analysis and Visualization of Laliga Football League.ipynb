```python
data = pd.read_csv(r"C:\Users\Admin\Desktop\Laliga.txt", skiprows=1)
column_names = ["Pos", "Team", "Seasons", "Points", "GamesPlayed", "GamesWon", "GamesDrawn", 
                "GamesLost", "GoalsFor", "GoalsAgainst", "Champion", "Runner-up", "Third", 
                "Fourth", "Fifth", "Sixth", "T", "Debut", "Since/LastApp", "BestPosition"]
data.columns = column_names
data.head()
```

## Data Cleaning and Preprocessing

```python
data = data.replace('-', 0)
data.head()
```

```python
data.isnull().sum()
```

## Statistical Analysis

```python
#Teams that started playing between 1930-1980
teams_between_1930_1980 = data[(data['Debut'] >= 1930) & (data['Debut'] <= 1980)]
print("Teams that started playing between 1930-1980:")
print(teams_between_1930_1980['Team'].tolist())
print()
```

```python
#Top 5 teams in terms of points
data['Points'] = pd.to_numeric(data['Points'])
top_5_teams = data.nlargest(5, 'Points')
print("Top 5 teams in terms of points:")
print(top_5_teams[['Team', 'Points']])
print()
```

```python
#Team which has maximum and minimum goal difference
data['GoalsFor'] = pd.to_numeric(data['GoalsFor'])
data['GoalsAgainst'] = pd.to_numeric(data['GoalsAgainst'])
def Goal_diff_count(row):
    return row['GoalsFor'] - row['GoalsAgainst']
data['GoalDifference'] = data.apply(Goal_diff_count, axis=1)
max_goal_diff_team = data.loc[data['GoalDifference'].idxmax()]
min_goal_diff_team = data.loc[data['GoalDifference'].idxmin()]
print("Team with maximum goal difference:")
print(max_goal_diff_team['Team'], ":", max_goal_diff_team['GoalDifference'])
print("Team with minimum goal difference:")
print(min_goal_diff_team['Team'], ":", min_goal_diff_team['GoalDifference'])
```

```python
#Top 5 teams with the highest Winning Percentage
data['GamesWon'] = pd.to_numeric(data['GamesWon'])
data['GamesPlayed'] = pd.to_numeric(data['GamesPlayed'])
data['WinningPercent'] = (data['GamesWon'] / data['GamesPlayed']) * 100
data['WinningPercent'] = data['WinningPercent'].fillna(0)
top_5_winning_percent = data.nlargest(5, 'WinningPercent')
print("Top 5 teams with the highest Winning Percentage:")
print(top_5_winning_percent[['Team', 'WinningPercent']])
print()
```

```python
#Teams based on their Best Position and the sum of their points for all positions
grouped_data = data.groupby('BestPosition')['Points'].sum().reset_index()
grouped_data.columns = ['Best Position', 'Total Points']
print("Total points for each Best Position:")
print(grouped_data)
```
